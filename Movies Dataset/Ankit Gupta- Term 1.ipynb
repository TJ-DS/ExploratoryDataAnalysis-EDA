```python
def bucket_Movi_length(Movi_length):
    if Movi_length > 30 and Movi_length < 50:
        return '30-50'
    elif Movi_length > 50 and Movi_length < 70:
        return '50-70'
    elif Movi_length > 70 and Movi_length < 90:
        return '70-90'
    elif Movi_length > 90 and Movi_length < 110:
        return '90-110'
    elif Movi_length > 110 and Movi_length < 130:
        return '110-130'
    elif Movi_length > 130 and Movi_length < 150:
        return '130-150'
    elif Movi_length > 150 and Movi_length < 170:
        return '150-170'
    elif Movi_length > 170 and Movi_length < 190:
        return '170-190'
    elif Movi_length > 190 and Movi_length < 210:
        return '190-210'


data['bucket_Movi_length'] = data.Movi_length.apply(bucket_Movi_length)
year_data = data.loc[data['Year'] == 2007]
group_year_lenght  = year_data.groupby(['bucket_Movi_length'])
group_year_lenght['bucket_Movi_length'].value_counts().plot.bar()

# Above plot It says, movie legth 110-130 count of movie is more.


```
