### 4.3 how many genres we are truly dealing with ?
Most of these films in the data set are part of multiple genres, We have to do spliting or something else. So we can entirely unique genres, without repeats

```python
import itertools
unique_Genre = MoviData['Genre'].unique()
individual_genres = []
for genre in unique_Genre:
    individual_genres.append(genre.split(','))
    
individual_genres = list(itertools.chain.from_iterable(individual_genres))
individual_genres = set(individual_genres)

individual_genres
```

```python
print('Number of movies in each genre: \n')

for genre in individual_genres:
    current_genre = MoviData['Genre'].str.contains(genre).fillna(False)
    plt.figure()
    plt.xlabel('Year')
    plt.ylabel('Number of Movies Made')
    plt.title(str(genre))
    MoviData[current_genre].Year.value_counts().sort_index().plot(kind='bar', color='r', alpha=0.5, rot=45)
    print(genre, len(MoviData[current_genre]))
```

### 4.3 genre percentages ?

__Since some films have multiple genres, these percentages won't add to 100%__

```python
genre_pcts = np.zeros(len(individual_genres))
i = 0
for genre in individual_genres:
    current_genre = MoviData['Genre'].str.contains(genre).fillna(False)
    pct = len(MoviData[current_genre]) / 1000 * 100
    genre_pcts[i] = pct
    i += 1
    
genre_pcts_df = pd.DataFrame(genre_pcts, index=individual_genres, columns=['Percent'])
genre_pcts_df= genre_pcts_df.sort_values(by=['Percent'])
genre_pcts_df
```

### 4.4 genre, determine revenue percentage 

```python
genre_revenue_pcts = np.zeros(len(individual_genres))
i = 0
for genre in individual_genres:
    current_genre = MoviData['Genre'].str.contains(genre).fillna(False)
    revenue_pct = MoviData[current_genre].xs('Revenue (Millions)', axis=1).sum() / MoviData['Revenue (Millions)'].sum() * 100
    genre_revenue_pcts[i] = revenue_pct
    i += 1
    print(genre, "==>" ,revenue_pct)
```

```python
genre_revenue_pcts_df = pd.DataFrame(genre_revenue_pcts, index=individual_genres, columns=['Percent'])
genre_revenue_pcts_df = genre_revenue_pcts_df.sort_values('Percent')
genre_revenue_pcts_df
```

```python
explode = (0.05, 0.05, 0.08, 0.1, 0.12)
genre_revenue_pcts_df.sort_values(by='Percent', ascending=False).head(5).plot.pie(legend=False, subplots=True, autopct='%.2f%%', figsize=(8,8),explode=explode)
plt.ylabel('')
plt.title('Percent of Total Revenue from Top 5 Genres', weight='bold', fontsize=16)
```

### 4. most active directors, 'most active' being defined as number of films with their name on it

```python
most_active_directors = MoviData['Director'].value_counts().head(10)
most_active_directors
most_active_directors.index
```

### 4. most active directors, how much revenu each of these top 10 directors

```python
director_revenue_totals = np.zeros(len(most_active_directors))
i = 0
for director in most_active_directors.index:
    current_director = MoviData['Director'].str.contains(director).fillna(False)
    director_film_revenue = MoviData[current_director].xs('Revenue (Millions)', axis=1).sum()
    director_revenue_totals[i] = director_film_revenue
    i += 1
    print(director, "===>", director_film_revenue)
    
director_revenue_totals_df = pd.DataFrame(director_revenue_totals, index=most_active_directors.index, columns=['Revenue (Millions)'])
director_revenue_totals_df.sort_values('Revenue (Millions)')
```

```python
# Taking the sum of the revenue of the films made by the top 10 directors, which contributed the most?
explode = np.linspace(0, 0.5, 10)
director_revenue_totals_df.sort_values(by='Revenue (Millions)', ascending=False).plot.pie(legend=False, subplots=True, autopct='%.2f%%', figsize=(8,8), explode=explode)
plt.ylabel('')
plt.title('Most Active Directors Revenue Contribution', weight='bold', fontsize=16)
```
